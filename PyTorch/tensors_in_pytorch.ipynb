## Mathematical Operations

### 1. Scalar operation

```python
x = torch.rand(2,2)
x
```

```python
# addition
x + 2
# substraction
x-2
# multiplication
x * 3
# division
x / 3
# int division
(x *100) // 3
# mod
((x *100) // 3) % 2
# power
x**2
```

### 2. Element wise operation

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
```

```python
# add
a + b
# sub
a - b
# multiply
a * b
# division
a / b
# power
a**b
# mod
a % b
```

```python
c = torch.tensor([1,-2,3,-4])
```

```python
# abs for absolute value
torch.abs(c)
```

```python
# negative
torch.neg(c) # 1 = -1, -2 = 2 etc
```

```python
# round
d = torch.tensor([1.9,2.3,3.7,4.4])

torch.round(d)
```

```python
# ceil

torch.ceil(d)
```

```python
# floor
torch.floor(d)
```

```python
# clamp
torch.clamp(d, min=2, max=3)
```

### 3. Reduction operation

```python
e = torch.randint(size=(2,3),low=0, high=10, dtype=torch.float32)
e # whole tensor in one digit using reduction operations
```

```python
# sum (sum the whole tensor)
torch.sum(e)
# sum along columns
torch.sum(e, dim=0)
# sum along rows
torch.sum(e, dim=1)
```

```python
# mean
torch.mean(e) # only works when dtype is float
torch.mean(e, dim=0)
```

```python
# median
torch.median(e)
```

```python
# max and min
torch.max(e)
torch.min(e)
```

```python
# product
torch.prod(e)
```

```python
# standard deviation
torch.std(e)
```

```python
# variance
torch.var(e)
```

```python
# argmax
torch.argmax(e) # position of the biggest argument
```

```python
# argmin
torch.argmin(e) # position of the smallest argument in the tensor
```

### 4. Matrix Operation

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)

print(f)
print(g)
```

```python
# matrix multiplication
torch.matmul(f,g)
```

```python
vector1 = torch.tensor([1,2])
vector2 = torch.tensor([3,4])

# dot product
torch.dot(vector1, vector2)
```

```python
# transpose
torch.transpose(f, 0, 1) # 0 to 1
```

```python
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
h
```

```python
# dererminant
torch.det(h)
```

```python
# inverse
torch.inverse(h)
```

### 5. Comparision Operations

```python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)
print(i)
print(j)
```

```python
# greater than
i > j
# less than
i < j
# equal to
i==j
# greater than equals to
i>=j
# less than equals to
i<=j
```

### 6. Special functions

```python
k = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
k
```

```python
# log
torch.log(k)
```

```python
# exponents
torch.exp(k)
```

```python
# sqrt
torch.sqrt(k)
```

```python
# sigmoid
torch.sigmoid(k)
```

```python
# softmax
torch.softmax(k,dim=0) # tensor must be float
```

```python
# relu
torch.relu(k)
```
